```python
results = {}
for i, group in tsne_df.groupby('group_str'):
    try:
        a = mk_fp(group)
    except Exception as error:
        print(error)
    else:
        results[i] = a
```

```python
from sklearn.manifold import TSNE
```

```python
def tsne_plots(a,b, label = None):
    summary = {}
    summary['label'] = label
    model = TSNE(learning_rate = 200, perplexity=a[label].shape[0]-1, random_state = 42)
    tsne_features = model.fit_transform(a[label])
    tsne_features = pd.DataFrame(tsne_features, index = a[label].index, columns = ['tsne_x', 'tsne_y'])
    tsne_df = pd.concat([tsne_features, b[label]], axis = 1)
    summary['tsne'] = tsne_df
    return summary
```

```python
tsne_lst = []
for k,v in results.items():
    if v.shape[0] > 1:
        tsne_lst.append(tsne_plots(results, phys_dict, label = k))
```

```python
tsne_all = []
for k,v in results.items():
    if v.shape[0] > 1:
        tsne_all.append(tsne_plots(results, universe_dict, label = k))
```

```python
tsne_all_dict = {}
for i in range(len(tsne_all)):
    a = tsne_all[i]['tsne'].reset_index()
    b = pd.melt(a, id_vars = ['index', 'tsne_x', 'tsne_y'], value_vars = ['final_centroid_tag', 'tsca_active_centroid_tag', 'MaxMin_tsca_active_picks','MaxMin_tsca_picks','tsca_centroid_tag', 'MaxMin_picks',  ])
    b = b.dropna(subset=['value']).drop_duplicates(subset = 'index', keep = 'first')
    tsne_all_dict[tsne_all[i]['label']] = a,b
```
